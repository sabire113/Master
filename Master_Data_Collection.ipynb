```python
import pandas as pd

# Load the dataset
file_path = "OSEBX_Monthly_Data_2015_2024.csv"
df = pd.read_csv(file_path)

# Ensure the Date column is in datetime format and sort by RIC and Date
df["Date"] = pd.to_datetime(df["Date"])
df = df.sort_values(by=["RIC", "Date"])

# Rename columns for consistency in calculations
df.rename(columns={
    "Price Close": "ClosePrice",
    "1 Month Total Return": "MonthlyReturn",
    "TRBC Economic Sector Name": "EconomicSector",
    "Company Market Cap": "MarketCap",
    "Common Shares - Outstanding - Total": "CommonSharesOutstanding",
    "Book Value Per Share": "BookValuePerShare",
    "Bid Price": "BidPrice",
    "Ask Price": "AskPrice",
    "Price Open": "OpenPrice",
    "Dividend yield": "DividendYield",
    "Earnings Per Share - Mean": "EarningsPerShare",
    "": "FirstTrade"
    
}, inplace=True)

# Fill missing EconomicSector values using forward fill (ffill) and backward fill (bfill)
df["EconomicSector"] = df.groupby("RIC")["EconomicSector"].ffill().bfill()

# Compute Monthly Returns if missing
if "MonthlyReturn" not in df.columns:
    df["MonthlyReturn"] = df.groupby("RIC")["ClosePrice"].pct_change()

# Define momentum and volatility periods
momentum_periods = [3, 6, 12]
volatility_periods = [3, 6, 12]

# Compute momentum (price change over k months)
for period in momentum_periods:
    df[f"Momentum_{period}M"] = df.groupby("RIC")["ClosePrice"].pct_change(periods=period)

# Compute volatility (std of monthly returns over k months)
for period in volatility_periods:
    df[f"Volatility_{period}M"] = df.groupby("RIC")["MonthlyReturn"].rolling(window=period, min_periods=1).std().reset_index(level=0, drop=True)

# Compute Bid-Ask Spread
df["BidAskSpread"] = (df["AskPrice"] - df["BidPrice"]) / df["ClosePrice"]

# Compute Turnover Ratio (Trading Volume / Common Shares Outstanding)
df["TurnoverRatio"] = df["Volume"] / df["CommonSharesOutstanding"]

# Save updated file
updated_file_path = "OSEBX_Monthly_Data_Updated.csv"
df.to_csv(updated_file_path, index=False)

print(f"Updated dataset saved as {updated_file_path}")
```

```python
import yfinance as yf
import pandas as pd
from fredapi import Fred
import requests

# Your FRED API Key
FRED_API_KEY = 'ced97fbd3eb985b14bb231f48c4d1e4d'
fred = Fred(api_key=FRED_API_KEY)

# Function to fetch data from FRED
def fetch_fred_data(series_id, start_date, end_date, resample_freq='M', fill_method=None):
    """
    Fetch time series data from FRED and resample it to the required frequency.
    """
    data = fred.get_series(series_id, observation_start=start_date, observation_end=end_date)
    if data is not None:
        df = pd.DataFrame(data, columns=[series_id])
        df.index = pd.to_datetime(df.index)
        df = df.resample(resample_freq).last()
        if fill_method:
            df = df.fillna(method=fill_method)
        return df
    return None

# Function to fetch Norges Bank 10-Year Government Zero-Coupon Bond Yield
def fetch_norges_bank_10y_yield(start_date, end_date):
    """Fetch Norges Bank 10-Year Government Zero-Coupon Bond Yield."""
    url = f"https://data.norges-bank.no/api/data/GOVT_ZEROCOUPON/M.10Y..?format=sdmx-json&startPeriod={start_date}&endPeriod={end_date}&locale=en"
    response = requests.get(url).json()
    
    if "data" not in response or "dataSets" not in response["data"]:
        print("Error: Unexpected API response structure")
        return pd.DataFrame()
    
    series = response["data"]["dataSets"][0].get("series", {})
    if not series:
        print("Error: No series data found in Norges Bank API response")
        return pd.DataFrame()
    
    series_key = list(series.keys())[0]  
    observations = series[series_key].get("observations", {})
    
    if not observations:
        print("Error: No observations found in Norges Bank API response")
        return pd.DataFrame()
    
    # Extracting time periods correctly
    structure = response["data"]["structure"]["dimensions"]["observation"]
    time_values = [d for d in structure if d["id"] == "TIME_PERIOD"]
    if time_values:
        time_values = time_values[0]["values"]
    else:
        print("Error: No time period information found in API response")
        return pd.DataFrame()
    
    time_index = {str(i): v["id"] for i, v in enumerate(time_values)}
    
    dates = []
    values = []
    for key, value in observations.items():
        if key in time_index:
            try:
                date = pd.to_datetime(time_index[key], format='%Y-%m')
                dates.append(date)
                values.append(value[0])
            except Exception as e:
                print(f"Skipping entry {key} due to error: {e}")
    
    if not dates:
        print("Error: No valid dates extracted from Norges Bank API response")
        return pd.DataFrame()
    
    df = pd.DataFrame({"Date": dates, "Norges Bank 10Y Yield": values})
    df.set_index("Date", inplace=True)
    df = df.resample("M").last()
    df = df.fillna(method='ffill')  # Forward fill missing values
    return df

# Define date range
start_date = '2015-01-01'
end_date = '2024-12-31'

### Fetch Yahoo Finance Data ###
brent_price = yf.download('BZ=F', start=start_date, end=end_date, interval='1mo')['Close']
brent_price.name = 'Brent Oil Price'

usd_nok = yf.download('USDNOK=X', start=start_date, end=end_date, interval='1mo')['Close']
usd_nok.name = 'USD/NOK Exchange Rate'

eur_nok = yf.download('EURNOK=X', start=start_date, end=end_date, interval='1mo')['Close']
eur_nok.name = 'EUR/NOK Exchange Rate'

osebx = yf.download('OSEBX.OL', start=start_date, end=end_date, interval='1mo')['Close']
osebx_returns = osebx.pct_change()
osebx_returns.name = 'OSEBX Returns'

### Fetch FRED Data ###
us_10y_rate = fetch_fred_data('DGS10', start_date, end_date)
us_10y_rate.rename(columns={'DGS10': 'US 10-Year Treasury Rate'}, inplace=True)

inflation_cpi = fetch_fred_data('CPIAUCSL', start_date, end_date)
inflation_cpi.rename(columns={'CPIAUCSL': 'Inflation (CPI)'}, inplace=True)

gdp_growth = fetch_fred_data('A191RL1Q225SBEA', start_date, end_date, resample_freq='M', fill_method='ffill')
gdp_growth.rename(columns={'A191RL1Q225SBEA': 'GDP Growth (Quarterly)'}, inplace=True)

### Fetch Norges Bank 10-Year Government Zero-Coupon Bond Yield ###
norges_bank_10y_yield = fetch_norges_bank_10y_yield(start_date, end_date)

### Ensure No Missing Macro Variables ###
macro_data = pd.concat(
    [brent_price, usd_nok, eur_nok, us_10y_rate, inflation_cpi, gdp_growth, osebx_returns, norges_bank_10y_yield], 
    axis=1
)
macro_data.index.name = "Date"
macro_data = macro_data.fillna(method='ffill')  # Forward-fill missing values
macro_data.to_csv("macroeconomic_data_full.csv")
print("All macroeconomic data saved to 'macroeconomic_data_full.csv'.")
```

```python
import requests
import pandas as pd

# Function to fetch Norwegian CPI with correct date parsing
def fetch_norwegian_cpi_fixed():
    url = "https://data.ssb.no/api/v0/en/table/05327/"
    query = {
        "query": [
            {
                "code": "Konsumgrp",
                "selection": {
                    "filter": "item",
                    "values": ["JA_TOTAL"]
                }
            },
            {
                "code": "ContentsCode",
                "selection": {
                    "filter": "item",
                    "values": ["KPIJustIndMnd"]
                }
            }
        ],
        "response": {
            "format": "json-stat2"
        }
    }

    response = requests.post(url, json=query)
    data = response.json()

    try:
        time_periods = list(data["dimension"]["Tid"]["category"]["label"].values())
        values = data["value"]

        # Convert 'YYYYMmm' format to 'YYYY-MM'
        formatted_dates = [pd.to_datetime(t[:4] + '-' + t[5:]) for t in time_periods]

        cpi_df = pd.DataFrame({"Date": formatted_dates, "NorwegianCPI": values})
        cpi_df.set_index("Date", inplace=True)
        return cpi_df
    except Exception as e:
        print(f"Error fetching Norwegian CPI: {e}")
        return pd.DataFrame()

# Fetch Norwegian CPI with the fixed function
norwegian_cpi_fixed = fetch_norwegian_cpi_fixed()

# Reload the existing macroeconomic data
file_path = "macroeconomic_data_full.csv"
macro_data = pd.read_csv(file_path, index_col=0, parse_dates=True)

# Merge Norwegian CPI into macro_data
macro_data_updated = macro_data.merge(norwegian_cpi_fixed, how="left", left_index=True, right_index=True)

# Define a dictionary to map old column names to new, simpler names
column_rename_mapping = {
    "Brent Oil Price": "BrentOil",
    "USD/NOK Exchange Rate": "USDNOK",
    "EUR/NOK Exchange Rate": "EURNOK",
    "US 10-Year Treasury Rate": "US10Y",
    "Inflation (CPI)": "USCPI",
    "GDP Growth (Quarterly)": "USGDPGrowth",
    "OSEBX Returns": "OSEBXReturns",
    "Norges Bank 10Y Yield": "NorgesBank10Y",
    "NorwegianCPI": "NorwegianCPI"
}

# Rename columns using the defined mapping
macro_data_updated.rename(columns=column_rename_mapping, inplace=True)

# Ensure 'NorwegianCPI' exists before forward-filling
if "NorwegianCPI" in macro_data_updated.columns:
    # Forward-fill missing Norwegian CPI values
    macro_data_updated["NorwegianCPI"].fillna(method="ffill", inplace=True)
    print("Norwegian CPI forward-filled successfully.")
else:
    print("Error: 'NorwegianCPI' column not found after merging.")

# Save the updated dataset
updated_file_path = "macroeconomic_data_full_fixed.csv"
macro_data_updated.to_csv(updated_file_path)

print(f"Updated dataset saved to '{updated_file_path}'.")
```
