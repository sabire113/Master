```python
# --- IMPORTS ---
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats.mstats import winsorize
from sklearn.metrics import r2_score, mean_squared_error
# *** Changed to HuberRegressor for GLM+H ***
from sklearn.linear_model import HuberRegressor
from sklearn.model_selection import ParameterGrid
import datetime
import warnings
import traceback
from collections import defaultdict
import os

# --- WARNINGS CONFIGURATION ---
warnings.filterwarnings("ignore", category=DeprecationWarning, module="pandas")
warnings.filterwarnings("ignore", category=FutureWarning)
warnings.filterwarnings("ignore", category=UserWarning, module="sklearn")
warnings.filterwarnings("ignore", category=RuntimeWarning, message="Mean of empty slice")
warnings.filterwarnings("ignore", category=RuntimeWarning, message="invalid value encountered in log")
warnings.filterwarnings("ignore", category=RuntimeWarning, message="Maximum number of iterations reached.*") # Ignore Huber convergence warnings
pd.options.mode.chained_assignment = None

# --------------------------------------------------------------------------
# FUNCTION DEFINITIONS (Steps 1 - 4 are identical to RF/GBRT-H)
# --------------------------------------------------------------------------

# ------------------------------------------------
# Step 1: Load and Prepare Dataset
# ------------------------------------------------
def load_prepare_data(file_path):
    """ Loads, cleans, calculates returns/market cap, handles dates/IDs. """
    print(f"Laster data fra: {file_path}")
    try:
        df = pd.read_csv(file_path, low_memory=False)
    except FileNotFoundError: print(f"FEIL: Fil '{file_path}' ikke funnet."); return None
    print(f"Data lastet inn. Form: {df.shape}")

    # Standardize essential column names (adjust if your CSV uses different names)
    date_col = 'Date' if 'Date' in df.columns else 'eom'
    id_col = 'Instrument' if 'Instrument' in df.columns else 'id'
    price_col = 'ClosePrice' if 'ClosePrice' in df.columns else 'prc'
    shares_col = 'CommonSharesOutstanding'
    rf_col = 'NorgesBank10Y'
    sector_col = 'EconomicSector'

    if date_col not in df.columns: print("FEIL: Dato-kolonne mangler."); return None
    if id_col not in df.columns: print("FEIL: Instrument-ID mangler."); return None
    if price_col not in df.columns: print(f"FEIL: Pris-kolonne ('{price_col}') mangler."); return None

    df = df.rename(columns={date_col: 'Date', id_col: 'Instrument'})
    df['Date'] = pd.to_datetime(df['Date'])
    df = df.sort_values(by=["Instrument", "Date"]).reset_index(drop=True)
    print("Dato konvertert og data sortert.")

    # Return Calculation
    df["MonthlyReturn"] = df.groupby("Instrument")[price_col].pct_change()
    df["MonthlyReturn"].fillna(0, inplace=True)
    df["MonthlyReturn"] = winsorize(df["MonthlyReturn"].values, limits=[0.01, 0.01])
    print("Månedlig avkastning ('MonthlyReturn') beregnet/winsorisert.")

    # Target Variable (Adjusted Return)
    if rf_col not in df.columns: df[rf_col] = 0; print(f"ADVARSEL: '{rf_col}' mangler, bruker 0.")
    # Handle if RF rate is in percent or decimal
    df["MonthlyRiskFreeRate_t"] = df[rf_col] / 12 / 100 if df[rf_col].abs().max() > 1 else df[rf_col] / 12
    df["TargetReturn_t"] = df["MonthlyReturn"] - df["MonthlyRiskFreeRate_t"] # Model target y
    print("Risikojustert avkastning ('TargetReturn_t') beregnet (modellens y).")

    # Next Month's RAW Return (for portfolio eval)
    df['NextMonthlyReturn_t+1'] = df.groupby('Instrument')['MonthlyReturn'].shift(-1)
    print("Neste måneds rå avkastning ('NextMonthlyReturn_t+1') beregnet.")

    # Market Cap
    if shares_col not in df.columns: print(f"FEIL: '{shares_col}' mangler for MarketCap."); return None
    df["MarketCap"] = df[price_col] * df[shares_col]
    df['MarketCap_orig'] = df['MarketCap'].copy() # Save before filling NA or logging
    df['MarketCap'] = df['MarketCap'].fillna(0)
    print("Markedsverdi ('MarketCap') beregnet.")

    # Sector Dummies (Optional)
    if sector_col in df.columns:
        df = pd.get_dummies(df, columns=[sector_col], prefix="Sector", dtype=int)
        print("Sektor dummy-variabler opprettet.")

    # Clean Column Names
    df.columns = df.columns.str.replace(" ", "_").str.replace("-", "_")
    print("Kolonnenavn renset.")

    # Log Transformation (on specific columns you defined)
    print("Log-transformerer spesifikke variabler...")
    vars_to_log = ["MarketCap", "BM", "ClosePrice", "Volume", "CommonSharesOutstanding"]
    vars_to_log = [v if v in df.columns else v.lower() for v in vars_to_log] # Check lower case
    vars_to_log = [v if v in df.columns else 'prc' if v == 'closeprice' else v for v in vars_to_log] # Map prc
    vars_to_log = [col for col in vars_to_log if col in df.columns] # Ensure columns exist after potential renames
    for var in vars_to_log:
        if var in df.columns: # Double check existence
             df[f"{var}_positive"] = df[var].where(df[var] > 1e-9, np.nan)
             df[f"log_{var}"] = np.log(df[f"{var}_positive"])
             log_median = df[f"log_{var}"].median()
             df[f"log_{var}"] = df[f"log_{var}"].fillna(log_median)
             if pd.isna(log_median): df[f"log_{var}"] = df[f"log_{var}"].fillna(0)
             df.drop(columns=[f"{var}_positive"], inplace=True)
             # print(f"  -> Log-transformert: {var} -> log_{var}") # Verbose
    print("Log-transformasjon fullført.")

    # Ensure MarketCap_orig exists if MarketCap was logged
    if 'MarketCap_orig' not in df.columns and 'MarketCap' in df.columns:
         df['MarketCap_orig'] = df['MarketCap'].copy()

    return df

# ------------------------------------------------
# Step 1.5: Rank Standardization
# ------------------------------------------------
def rank_standardize_features(df, features_to_standardize):
    """ Performs rank standardization cross-sectionally to [-1, 1]. """
    print(f"Rank standardiserer {len(features_to_standardize)} features...")
    if 'Date' not in df.columns: print("FEIL: 'Date' mangler."); return df
    features_present = [f for f in features_to_standardize if f in df.columns]
    if len(features_present) < len(features_to_standardize):
        missing = [f for f in features_to_standardize if f not in features_present]
        print(f"  ADVARSEL: Følgende features manglet for standardisering: {missing}")
    if not features_present: print("  Ingen features å standardisere."); return df

    def rank_transform(x): return x.rank(pct=True) * 2 - 1
    try:
        ranked_cols = df.groupby('Date')[features_present].transform(rank_transform)
        df[features_present] = ranked_cols
    except Exception as e:
        print(f"  FEIL under rank standardisering (transform): {e}. Prøver apply.")
        try:
            df_std = df.set_index('Date')
            for col in features_present: df_std[col] = df_std.groupby(level=0)[col].rank(pct=True) * 2 - 1
            df = df_std.reset_index()
        except Exception as e2:
            print(f"  FEIL: Også alternativ standardisering feilet: {e2}"); return df # Return original on failure

    print("Rank standardisering fullført.")
    return df

# ------------------------------------------------
# Step 2: Define Feature Sets (Auto-detects from df)
# ------------------------------------------------
def define_features(df):
    """ Identifies numeric predictor features from DataFrame 'df'. """
    print("Identifiserer numeriske features...")
    if df is None or df.empty: print("  FEIL: DataFrame er tom."); return []

    numeric_cols = df.select_dtypes(include=np.number).columns.tolist()
    print(f"  Funnet {len(numeric_cols)} numeriske kolonner totalt.")

    cols_to_exclude = ['Instrument', 'Date', 'level_0', 'index', 'TargetReturn_t', 'NextMonthlyReturn_t+1', 'MonthlyReturn', 'MonthlyRiskFreeRate_t', 'MarketCap_orig', 'Year', 'MonthYear', 'rank', 'DecileRank', 'eq_weights', 'me_weights']
    cols_to_exclude.extend([col for col in df.columns if 'return_stock' in col or 'return_portfolio' in col])
    log_cols = [col for col in numeric_cols if col.startswith('log_')]
    originals_of_log = [col.replace('log_','') for col in log_cols]
    originals_to_exclude = [orig for orig in originals_of_log if orig in df.columns]; cols_to_exclude.extend(originals_to_exclude)
    if 'log_MarketCap' in log_cols: cols_to_exclude.append('MarketCap')
    if 'log_ClosePrice' in log_cols or 'log_prc' in log_cols: cols_to_exclude.extend(['ClosePrice','prc'])
    if 'log_Volume' in log_cols: cols_to_exclude.append('Volume')
    if 'log_CommonSharesOutstanding' in log_cols: cols_to_exclude.append('CommonSharesOutstanding')
    if 'log_BM' in log_cols: cols_to_exclude.append('BM')
    cols_to_exclude = list(set(cols_to_exclude))

    potential_features = [col for col in numeric_cols if col not in cols_to_exclude]
    final_features = []
    for col in potential_features:
         if df[col].nunique(dropna=False) > 1 and df[col].std(ddof=0) > 1e-9: final_features.append(col)
    final_features = sorted(list(set(final_features)))
    print(f"  Identifisert {len(final_features)} features for bruk i modellen.")
    return final_features

# ------------------------------------------------
# Step 3: Handle Missing / Infinite Values
# ------------------------------------------------
def clean_data(df, numeric_features_to_impute, essential_cols_for_dropna, target="TargetReturn_t"):
    """ Handles inf/NaN after standardization. Filters essential cols and MarketCap_orig <= 0. """
    print("Starter datarensing (missing/inf)...")
    initial_rows = len(df)
    features_present = [f for f in numeric_features_to_impute if f in df.columns]
    if not features_present: print("  Ingen features for imputering.");
    else:
        inf_mask = df[features_present].isin([np.inf, -np.inf])
        if inf_mask.any().any(): print(f"  Erstatter inf med NaN..."); df[features_present] = df[features_present].replace([np.inf, -np.inf], np.nan)
        nan_counts_before = df[features_present].isnull().sum(); medians = df[features_present].median(); df[features_present] = df[features_present].fillna(medians); nan_counts_after = df[features_present].isnull().sum()
        imputed_cols = (nan_counts_before - nan_counts_after); print(f"  NaNs imputert med median i {imputed_cols[imputed_cols > 0].count()} kolonner.")
        if medians.isnull().any(): cols_nan_median = medians[medians.isnull()].index.tolist(); print(f"  ADVARSEL: Median NaN for: {cols_nan_median}. Fylte med 0."); df[cols_nan_median] = df[cols_nan_median].fillna(0)

    essential_cols_present = [col for col in essential_cols_for_dropna if col in df.columns]; rows_before_dropna = len(df)
    for col in [target,'NextMonthlyReturn_t+1','MarketCap_orig']:
         if col in df.columns and col not in essential_cols_present: essential_cols_present.append(col)
    unique_essential_cols = sorted(list(set(essential_cols_present)))
    if unique_essential_cols: df = df.dropna(subset=unique_essential_cols); print(f"  Fjernet {rows_before_dropna - len(df)} rader pga. NaN i: {unique_essential_cols}")

    mc_orig_col = 'MarketCap_orig'
    if mc_orig_col in df.columns:
        rows_before_mc_filter = len(df); df = df[df[mc_orig_col] > 0]; print(f"  Fjernet {rows_before_mc_filter - len(df)} rader der {mc_orig_col} <= 0.")
    else: print(f"  Advarsel: '{mc_orig_col}' ikke funnet.")
    final_rows = len(df); print(f"Datarensing fullført. Form: {df.shape}. Fjernet totalt {initial_rows - final_rows} rader.")
    if df.empty: print("FEIL: Ingen data igjen etter rensing.")
    return df

# ------------------------------------------------
# Step 4: Rolling Window Splits
# ------------------------------------------------
def get_rolling_splits(df, train_period, val_period, test_period):
    """ Creates rolling window splits based on MonthYear periods. """
    if "Date" not in df.columns: raise ValueError("'Date'-kolonnen mangler.")
    temp_col = False
    if 'MonthYear' not in df.columns: df['MonthYear'] = df["Date"].dt.to_period('M'); temp_col = True
    unique_dates = sorted(df["MonthYear"].unique()); n_unique_months=len(unique_dates); total_window_months = train_period+val_period+test_period
    print(f"Unike måneder: {n_unique_months}, Total vindu: {total_window_months} mnd")
    num_windows = n_unique_months - total_window_months + 1
    if num_windows <= 0:
        if temp_col: df.drop(columns=['MonthYear'], inplace=True)
        raise ValueError(f"Ikke nok data ({n_unique_months}) for vinduer ({total_window_months}).")
    print(f"Genererer {num_windows} rullerende vinduer...")
    for i in range(num_windows):
        train_start=unique_dates[i]; train_end=unique_dates[i+train_period-1]; val_start=unique_dates[i+train_period]; val_end=unique_dates[i+train_period+val_period-1]; test_start=unique_dates[i+train_period+val_period]; test_end=unique_dates[i+total_window_months-1]
        train_idx=df[(df['MonthYear'] >= train_start) & (df['MonthYear'] <= train_end)].index; val_idx=df[(df['MonthYear'] >= val_start) & (df['MonthYear'] <= val_end)].index; test_idx=df[(df['MonthYear'] >= test_start) & (df['MonthYear'] <= test_end)].index
        train_dates=df.loc[train_idx, "Date"].agg(['min','max']) if not train_idx.empty else None; val_dates=df.loc[val_idx, "Date"].agg(['min','max']) if not val_idx.empty else None; test_dates=df.loc[test_idx, "Date"].agg(['min','max']) if not test_idx.empty else None
        yield train_idx, val_idx, test_idx, train_dates, val_dates, test_dates
    if temp_col: df.drop(columns=['MonthYear'], inplace=True) # Clean up temp col


# --------------------------------------------------------------------------
# Step 5: Run GLM with Huber Loss (HuberRegressor) on a Single Rolling Window
# --------------------------------------------------------------------------
def run_glm_huber_on_window(X_train, y_train, X_val, y_val, X_test, y_test, param_grid=None):
    """ Trains HuberRegressor with tuning for alpha and epsilon. Returns metrics & optimal alpha. """
    model = None; optim_param_found = None; optimal_alpha = np.nan
    preds_oos = np.full(y_test.shape[0], np.nan);
    preds_combined_is = np.full(y_train.shape[0] + y_val.shape[0], np.nan)
    r2_oos, mse_oos, sharpe_oos, r2_is, sharpe_is = (np.nan,) * 5

    if param_grid is None: # Default GLM-H Grid
         # Alpha (regularization strength), Epsilon (Huber threshold)
         param_grid = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0],
                       'epsilon': [1.1, 1.35, 1.5, 1.75, 2.0]}
                       # Add 'max_iter': [300] or higher if convergence issues arise

    grid = list(ParameterGrid(param_grid)); mse_results = np.full(len(grid), np.inf); best_mse = np.inf
    # print(f"    Tuning GLM-H ({len(grid)} combos)...") # Less verbose
    if X_val.shape[0] < 2: print("    Val set too small. Skipping."); return preds_oos, r2_oos, mse_oos, sharpe_oos, r2_is, sharpe_is, model, optimal_alpha

    for i, params in enumerate(grid): # Tuning loop
        try:
             # Add max_iter if needed, e.g., max_iter=300
             glm_val = HuberRegressor(fit_intercept=True, **params, max_iter=300) # Increased max_iter
             glm_val.fit(X_train, y_train)
             y_val_pred = glm_val.predict(X_val)
             current_mse = mean_squared_error(y_val, y_val_pred)
             if not np.isnan(current_mse) and current_mse < best_mse: best_mse = current_mse; optim_param_found = params
        except Exception as e:
            # print(f"      Tuning error for params {params}: {e}") # Optional debug
            continue # Ignore errors during tuning (e.g., convergence)

    if optim_param_found is None: print("    GLM-H Tuning failed."); return preds_oos, r2_oos, mse_oos, sharpe_oos, r2_is, sharpe_is, model, optimal_alpha
    optimal_alpha = optim_param_found.get('alpha', np.nan); # print(f"    Optimal GLM-H params: {optim_param_found}") # Less verbose

    try: # Final Training
        # print("    Training final GLM-H..."); # Less verbose
        X_train_val = np.vstack((X_train, X_val)); y_train_val = np.concatenate((y_train, y_val))
        # Add max_iter=300 here too
        model = HuberRegressor(fit_intercept=True, **optim_param_found, max_iter=300)
        model.fit(X_train_val, y_train_val)
        # OOS Eval
        preds_oos = model.predict(X_test); nan_preds_oos = np.isnan(preds_oos); preds_oos[nan_preds_oos] = 0
        y_test_valid = y_test[~nan_preds_oos]; preds_oos_valid = preds_oos[~nan_preds_oos]
        if len(preds_oos_valid) > 1:
            r2_oos = r2_score(y_test_valid, preds_oos_valid); mse_oos = mean_squared_error(y_test_valid, preds_oos_valid)
            pred_std_oos = np.std(preds_oos_valid); sharpe_oos = np.mean(preds_oos_valid)/pred_std_oos if pred_std_oos > 1e-9 else np.nan
        # IS Eval
        preds_combined_is = model.predict(X_train_val); nan_preds_is = np.isnan(preds_combined_is); preds_combined_is[nan_preds_is] = 0
        y_train_val_valid = y_train_val[~nan_preds_is]; preds_is_valid = preds_combined_is[~nan_preds_is]
        if len(preds_is_valid) > 1:
            r2_is = r2_score(y_train_val_valid, preds_is_valid)
            pred_std_is = np.std(preds_is_valid); sharpe_is = np.mean(preds_is_valid)/pred_std_is if pred_std_is > 1e-9 else np.nan
    except Exception as e:
        print(f"  FEIL final GLM-H fit/predict: {e}"); model = None; optimal_alpha = np.nan
        preds_oos.fill(np.nan); r2_oos, mse_oos, sharpe_oos, r2_is, sharpe_is = (np.nan,) * 5

    # Return optimal_alpha instead of optimal_depth for consistency in signature
    return preds_oos, r2_oos, mse_oos, sharpe_oos, r2_is, sharpe_is, model, optimal_alpha


# -----------------------------------------------------------------------------
# Step 6.5: Detailed Portfolio Analysis Function (Identical to RF/GBRT-H)
# -----------------------------------------------------------------------------
def MDD(returns):
    """ Calculates Maximum Drawdown using arithmetic returns for NAV. """
    returns = pd.Series(returns).fillna(0)
    if returns.empty: return np.nan
    nav = (1 + returns).cumprod() * 100; hwm = nav.cummax(); dd = nav / hwm - 1
    return dd.min() if not dd.empty else np.nan

def perform_detailed_portfolio_analysis(results_df, original_df_subset, benchmark_file=None, filter_small_caps=False, model_name_label="GLM-H"): # <-- Default Label Changed
    """ Performs detailed portfolio analysis, generates tables and plots for the given model label. """
    print("\n--- Starter Detaljert Porteføljeanalyse ---")
    pred_col = f'yhat_{model_name_label.lower().replace("-","_")}' # e.g., yhat_glm_h
    ew_table, vw_table, ew_chart, vw_chart, ew_long_chart, vw_long_chart = (pd.DataFrame(),)*6

    if pred_col not in results_df.columns: print(f"FEIL: Kolonne '{pred_col}' mangler."); return ew_table, vw_table, ew_chart, vw_chart, ew_long_chart, vw_long_chart
    required_cols = ['Date', 'Instrument', 'MarketCap_orig', 'NextMonthlyReturn_t+1', 'MonthlyReturn', 'MonthlyRiskFreeRate_t']
    if not all(c in original_df_subset.columns for c in required_cols): print("FEIL: Mangler kolonner i original subset."); return ew_table, vw_table, ew_chart, vw_chart, ew_long_chart, vw_long_chart
    portfolio_data = pd.merge(results_df[['Date', 'Instrument', 'TargetReturn_t', pred_col]], original_df_subset[required_cols], on=['Date', 'Instrument'], how='inner')
    portfolio_data = portfolio_data.rename(columns={'TargetReturn_t': 'y_true_t', pred_col: 'yhat_t+1', 'MarketCap_orig': 'me', 'NextMonthlyReturn_t+1': 'ret_t+1'})
    portfolio_data['MonthYear'] = portfolio_data['Date'].dt.to_period('M')
    monthly_rf_map = portfolio_data.groupby('MonthYear')['MonthlyRiskFreeRate_t'].mean().shift(-1)
    portfolio_data['NextMonthRiskFreeRate_t+1'] = portfolio_data['MonthYear'].map(monthly_rf_map)
    cols_for_eval_dropna = ['yhat_t+1', 'ret_t+1', 'me', 'NextMonthRiskFreeRate_t+1']
    portfolio_data = portfolio_data.dropna(subset=cols_for_eval_dropna)
    if portfolio_data.empty: print("  FEIL: Ingen data igjen for analyse."); return ew_table, vw_table, ew_chart, vw_chart, ew_long_chart, vw_long_chart
    portfolio_data['excess_ret_t+1'] = portfolio_data['ret_t+1'] - portfolio_data['NextMonthRiskFreeRate_t+1']

    print("  Sorterer i desiler og beregner vekter...")
    monthly_data_dict = {}; all_decile_dfs = {i: [] for i in range(10)}
    unique_months = sorted(portfolio_data['MonthYear'].unique())
    for month in unique_months:
        monthly_df = portfolio_data[portfolio_data['MonthYear'] == month].copy()
        if filter_small_caps:
            if 'me' in monthly_df.columns and len(monthly_df) > 10: mc_cutoff = monthly_df['me'].quantile(0.10); monthly_df = monthly_df[monthly_df['me'] >= mc_cutoff].copy()
        if len(monthly_df) < 10: continue
        monthly_df = monthly_df.sort_values('yhat_t+1')
        try:
            monthly_df['rank'] = monthly_df['yhat_t+1'].rank(method='first'); monthly_df['DecileRank'] = pd.qcut(monthly_df['rank'], 10, labels=False, duplicates='drop')
            if monthly_df['DecileRank'].nunique() < 10: continue
        except ValueError: continue
        monthly_df = monthly_df.drop(columns=['rank']); monthly_df["eq_weights"] = 1 / monthly_df.groupby('DecileRank')["Instrument"].transform('size'); monthly_df["me_weights"] = monthly_df["me"] / monthly_df.groupby('DecileRank')["me"].transform('sum'); monthly_df["me_weights"] = monthly_df["me_weights"].fillna(0); monthly_data_dict[month] = monthly_df
        for decile_rank, group_df in monthly_df.groupby('DecileRank'): all_decile_dfs[decile_rank].append(group_df)
    decile_portfolios = {j: pd.concat(all_decile_dfs[j], ignore_index=True) if all_decile_dfs[j] else pd.DataFrame() for j in range(10)}
    if not any(not df.empty for df in decile_portfolios.values()): print("  FEIL: Ingen desilporteføljer konstruert."); return ew_table, vw_table, ew_chart, vw_chart, ew_long_chart, vw_long_chart

    print("  Beregner desil-metrikker...")
    decile_results = []; monthly_agg_data = {}
    for j in range(10):
        rank_df = decile_portfolios[j];
        if rank_df.empty: continue
        rank_df['excess_return_stock_ew']=rank_df["excess_ret_t+1"]*rank_df["eq_weights"]; rank_df['excess_return_stock_vw']=rank_df["excess_ret_t+1"]*rank_df["me_weights"]
        rank_df['pred_excess_return_stock_ew']=rank_df["yhat_t+1"]*rank_df["eq_weights"]; rank_df['pred_excess_return_stock_vw']=rank_df["yhat_t+1"]*rank_df["me_weights"]
        rank_df['return_stock_ew']=rank_df["ret_t+1"]*rank_df["eq_weights"]; rank_df['return_stock_vw']=rank_df["ret_t+1"]*rank_df["me_weights"]
        monthly_rank_j = rank_df.groupby('MonthYear').agg(excess_return_portfolio_ew=('excess_return_stock_ew','sum'), excess_return_portfolio_vw=('excess_return_stock_vw','sum'), pred_excess_return_portfolio_ew=('pred_excess_return_stock_ew','sum'), pred_excess_return_portfolio_vw=('pred_excess_return_stock_vw','sum'), return_portfolio_ew=('return_stock_ew','sum'), return_portfolio_vw=('return_stock_vw','sum')).reset_index()
        monthly_rank_j['DecileRank'] = j; monthly_agg_data[j] = monthly_rank_j
        ew_mean_ret=monthly_rank_j["excess_return_portfolio_ew"].mean(); vw_mean_ret=monthly_rank_j["excess_return_portfolio_vw"].mean()
        ew_mean_pred=monthly_rank_j["pred_excess_return_portfolio_ew"].mean(); vw_mean_pred=monthly_rank_j["pred_excess_return_portfolio_vw"].mean()
        std_ew_raw=monthly_rank_j["return_portfolio_ew"].std(); std_vw_raw=monthly_rank_j["return_portfolio_vw"].std()
        sharpe_ew=(ew_mean_ret/std_ew_raw)*np.sqrt(12) if std_ew_raw>1e-9 else np.nan; sharpe_vw=(vw_mean_ret/std_vw_raw)*np.sqrt(12) if std_vw_raw>1e-9 else np.nan
        decile_results.append({'DecileRank':j,'ew_mean_pred':ew_mean_pred,'ew_mean_ret':ew_mean_ret,'std_ew_ret':std_ew_raw,'sharpe_ew':sharpe_ew,'vw_mean_pred':vw_mean_pred,'vw_mean_ret':vw_mean_ret,'std_vw_ret':std_vw_raw,'sharpe_vw':sharpe_vw})

    zeronet_monthly = pd.DataFrame()
    if 0 in monthly_agg_data and 9 in monthly_agg_data and not monthly_agg_data[0].empty and not monthly_agg_data[9].empty:
        long_monthly=monthly_agg_data[9].set_index('MonthYear'); short_monthly=monthly_agg_data[0].set_index('MonthYear'); common_index=long_monthly.index.intersection(short_monthly.index)
        if not common_index.empty:
             zeronet_monthly = long_monthly.loc[common_index].subtract(short_monthly.loc[common_index], fill_value=0).rename(columns=lambda x: x+'_HL').drop(columns=['DecileRank_HL'],errors='ignore').reset_index()
             ew_mean_ret_hl=zeronet_monthly["excess_return_portfolio_ew_HL"].mean(); vw_mean_ret_hl=zeronet_monthly["excess_return_portfolio_vw_HL"].mean(); ew_mean_pred_hl=zeronet_monthly["pred_excess_return_portfolio_ew_HL"].mean(); vw_mean_pred_hl=zeronet_monthly["pred_excess_return_portfolio_vw_HL"].mean()
             std_ew_raw_hl=zeronet_monthly["return_portfolio_ew_HL"].std(); std_vw_raw_hl=zeronet_monthly["return_portfolio_vw_HL"].std()
             sharpe_ew_hl=(ew_mean_ret_hl/std_ew_raw_hl)*np.sqrt(12) if std_ew_raw_hl>1e-9 else np.nan; sharpe_vw_hl=(vw_mean_ret_hl/std_vw_raw_hl)*np.sqrt(12) if std_vw_raw_hl>1e-9 else np.nan
             decile_results.append({'DecileRank':'H-L','ew_mean_pred':ew_mean_pred_hl,'ew_mean_ret':ew_mean_ret_hl,'std_ew_ret':std_ew_raw_hl,'sharpe_ew':sharpe_ew_hl,'vw_mean_pred':vw_mean_pred_hl,'vw_mean_ret':vw_mean_ret_hl,'std_vw_ret':std_vw_raw_hl,'sharpe_vw':sharpe_vw_hl})
        else: print("  ADVARSEL: Ingen overlappende måneder for H-L.")
    else: print("  ADVARSEL: Kan ikke beregne H-L.");

    if not decile_results: print("  FEIL: Ingen desilresultater."); return (None,)*6
    results_summary_df = pd.DataFrame(decile_results).set_index('DecileRank')
    def format_decile_table(summary_df, weight_scheme):
        cols_map={'ew_mean_pred':'Pred','ew_mean_ret':'Real','std_ew_ret':'Std','sharpe_ew':'Sharpe'} if weight_scheme=='EW' else {'vw_mean_pred':'Pred','vw_mean_ret':'Real','std_vw_ret':'Std','sharpe_vw':'Sharpe'}
        sub_df=summary_df[[k for k in cols_map.keys() if k in summary_df.columns]].rename(columns=cols_map)
        for col in ['Pred','Real','Std']:
            if col in sub_df.columns: sub_df[col]=(sub_df[col]*100).map('{:.2f}%'.format).replace('nan%','N/A')
        if 'Sharpe' in sub_df.columns: sub_df['Sharpe']=sub_df['Sharpe'].map('{:.2f}'.format).replace('nan','N/A')
        def map_index(x):
            if x==0: return 'Low (L)';
            if x==9: return 'High (H)';
            if x=='H-L': return 'H-L'
            try: return str(int(x)+1)
            except: return str(x)
        sub_df.index=sub_df.index.map(map_index); desired_order=['Low (L)','2','3','4','5','6','7','8','9','High (H)','H-L']; sub_df=sub_df.reindex([idx for idx in desired_order if idx in sub_df.index])
        return sub_df[[col for col in ['Pred','Real','Std','Sharpe'] if col in sub_df.columns]]
    ew_table = format_decile_table(results_summary_df, 'EW'); vw_table = format_decile_table(results_summary_df, 'VW')
    print(f"\n--- Ytelsestabell ({model_name_label}) - EW ---"); print(ew_table); print(f"\n--- Ytelsestabell ({model_name_label}) - VW ---"); print(vw_table)

    print("\n--- Ytterligere H-L Porteføljeanalyse ---")
    turnover_ew, turnover_vw, maxDD_ew, maxDD_vw, max_loss_ew, max_loss_vw = (np.nan,) * 6
    if not zeronet_monthly.empty and monthly_data_dict:
        all_months_weights = pd.concat(monthly_data_dict.values(), ignore_index=True)
        if 0 in all_months_weights['DecileRank'].unique() and 9 in all_months_weights['DecileRank'].unique():
             long_weights = all_months_weights[all_months_weights['DecileRank']==9][['MonthYear','Instrument','eq_weights','me_weights']]
             short_weights = all_months_weights[all_months_weights['DecileRank']==0][['MonthYear','Instrument','eq_weights','me_weights']]; short_weights[['eq_weights','me_weights']]*=-1
             hl_weights = pd.concat([long_weights,short_weights]).sort_values(['Instrument','MonthYear'])
             hl_weights['eq_weights_lead1'] = hl_weights.groupby('Instrument')['eq_weights'].shift(-1).fillna(0); hl_weights['me_weights_lead1'] = hl_weights.groupby('Instrument')['me_weights'].shift(-1).fillna(0)
             hl_weights['abs_diff_ew'] = abs(hl_weights['eq_weights_lead1']-hl_weights['eq_weights']); hl_weights['abs_diff_vw'] = abs(hl_weights['me_weights_lead1']-hl_weights['me_weights'])
             last_month = hl_weights['MonthYear'].max(); monthly_turnover = hl_weights[hl_weights['MonthYear'] != last_month].groupby('MonthYear').agg(sum_abs_ew=('abs_diff_ew','sum'),sum_abs_vw=('abs_diff_vw','sum'))
             if not monthly_turnover.empty: turnover_ew=monthly_turnover['sum_abs_ew'].mean(); turnover_vw=monthly_turnover['sum_abs_vw'].mean(); print(f"  Avg Turnover (H-L): EW={turnover_ew*100:.2f}%, VW={turnover_vw*100:.2f}%")
        else: print("  Kan ikke beregne turnover (mangler desil 0/9).")
        maxDD_ew=MDD(zeronet_monthly['excess_return_portfolio_ew_HL']); maxDD_vw=MDD(zeronet_monthly['excess_return_portfolio_vw_HL']); print(f"  Max Drawdown (H-L): EW={abs(maxDD_ew)*100:.2f}%, VW={abs(maxDD_vw)*100:.2f}%")
        max_loss_ew=zeronet_monthly['excess_return_portfolio_ew_HL'].min(); max_loss_vw=zeronet_monthly['excess_return_portfolio_vw_HL'].min(); print(f"  Max 1M Loss (H-L): EW={abs(max_loss_ew)*100:.2f}%, VW={abs(max_loss_vw)*100:.2f}%")
    else: print("  Kan ikke utføre H-L turnover/drawdown.")

    benchmark_name = "OSEBX"; print(f"\n--- Sammenligning med Benchmark ({benchmark_name}) ---")
    bench_data_loaded = False; ew_chart = pd.DataFrame(); vw_chart = pd.DataFrame(); ew_long_chart=pd.DataFrame(); vw_long_chart=pd.DataFrame();
    mean_ret_bench_pct, std_bench_pct, sr_bench, mdd_bench_pct = (np.nan,) * 4
    if benchmark_file and os.path.exists(benchmark_file) and not zeronet_monthly.empty:
        try:
            bench = pd.read_csv(benchmark_file, parse_dates=['Date']); bench=bench.sort_values('Date'); bench['MonthYear']=bench['Date'].dt.to_period('M'); bench=bench.drop_duplicates(subset=['MonthYear'],keep='last'); bench['monthly_return_bench']=bench['Close'].pct_change()
            avg_monthly_rf = portfolio_data.groupby('MonthYear')['MonthlyRiskFreeRate_t'].mean()
            bench = pd.merge(bench[['MonthYear','monthly_return_bench']],avg_monthly_rf.reset_index(), on='MonthYear',how='left'); bench['monthly_excess_return_bench']=bench['monthly_return_bench']-bench['MonthlyRiskFreeRate_t']
            bench_zeronet=pd.merge(bench[['MonthYear','monthly_return_bench','monthly_excess_return_bench']],zeronet_monthly,on='MonthYear',how='inner').dropna()
            if not bench_zeronet.empty:
                bench_data_loaded = True
                mean_ret_bench_pct=bench_zeronet["monthly_excess_return_bench"].mean()*100; std_bench_pct=bench_zeronet["monthly_excess_return_bench"].std()*100
                sr_bench=(bench_zeronet["monthly_excess_return_bench"].mean()/bench_zeronet["monthly_return_bench"].std())*np.sqrt(12) if bench_zeronet["monthly_return_bench"].std()>1e-9 else np.nan
                mdd_bench_pct=abs(MDD(bench_zeronet["monthly_excess_return_bench"]))*100
                hl_res=results_summary_df.loc['H-L'] if 'H-L' in results_summary_df.index else pd.Series(dtype=float)
                active_ret_ew=bench_zeronet["excess_return_portfolio_ew_HL"]-bench_zeronet["monthly_excess_return_bench"]; active_ret_vw=bench_zeronet["excess_return_portfolio_vw_HL"]-bench_zeronet["monthly_excess_return_bench"]
                mean_active_ret_ew=active_ret_ew.mean(); mean_active_ret_vw=active_ret_vw.mean()
                ir_ew=(mean_active_ret_ew/active_ret_ew.std())*np.sqrt(12) if active_ret_ew.std()>1e-9 else np.nan; ir_vw=(mean_active_ret_vw/active_ret_vw.std())*np.sqrt(12) if active_ret_vw.std()>1e-9 else np.nan
                index_perf = ["Mean Excess Return [%]",'Std Excess Return [%]', "Ann. Sharpe Ratio", "Max Drawdown [%]","Avg Monthly Turnover [%]", "Mean Active Return","Ann. Information Ratio"]
                ew_chart = pd.DataFrame({benchmark_name:[mean_ret_bench_pct,std_bench_pct,sr_bench,mdd_bench_pct,0,0,0],f'{model_name_label} H-L':[hl_res.get('ew_mean_ret',np.nan)*100,hl_res.get('std_ew_ret',np.nan)*100,hl_res.get('sharpe_ew',np.nan),abs(maxDD_ew)*100,turnover_ew*100,mean_active_ret_ew*100,ir_ew]},index=index_perf)
                vw_chart = pd.DataFrame({benchmark_name:[mean_ret_bench_pct,std_bench_pct,sr_bench,mdd_bench_pct,0,0,0],f'{model_name_label} H-L':[hl_res.get('vw_mean_ret',np.nan)*100,hl_res.get('std_vw_ret',np.nan)*100,hl_res.get('sharpe_vw',np.nan),abs(maxDD_vw)*100,turnover_vw*100,mean_active_ret_vw*100,ir_vw]},index=index_perf)
                print(f"\n--- Risk-Adjusted Performance (H-L vs {benchmark_name}, EW) ---"); print(ew_chart.round(3)); print(f"\n--- Risk-Adjusted Performance (H-L vs {benchmark_name}, VW) ---"); print(vw_chart.round(3))
            else: print("  Ingen overlappende H-L data med benchmark.")
        except Exception as e: print(f"  FEIL ved benchmark analyse: {e}")
    else: print(f"  Benchmark data ikke tilgjengelig/spesifisert eller H-L mangler.")

    print("\n--- Long-Only (Top Decile) Porteføljeanalyse ---")
    rank_9_df = decile_portfolios.get(9, pd.DataFrame()); turnover_ew_long, turnover_vw_long, maxDD_ew_long, maxDD_vw_long, max_loss_ew_long, max_loss_vw_long = (np.nan,) * 6
    if not rank_9_df.empty:
        long_weights = rank_9_df[['MonthYear','Instrument','eq_weights','me_weights']].copy().sort_values(['Instrument','MonthYear'])
        long_weights['eq_weights_lead1']=long_weights.groupby('Instrument')['eq_weights'].shift(-1).fillna(0); long_weights['me_weights_lead1']=long_weights.groupby('Instrument')['me_weights'].shift(-1).fillna(0); long_weights['abs_diff_ew']=abs(long_weights['eq_weights_lead1']-long_weights['eq_weights']); long_weights['abs_diff_vw']=abs(long_weights['me_weights_lead1']-long_weights['me_weights'])
        monthly_turnover_long=long_weights[long_weights['MonthYear']!=long_weights['MonthYear'].max()].groupby('MonthYear').agg(sum_abs_ew=('abs_diff_ew','sum'),sum_abs_vw=('abs_diff_vw','sum'))
        if not monthly_turnover_long.empty: turnover_ew_long=monthly_turnover_long['sum_abs_ew'].mean(); turnover_vw_long=monthly_turnover_long['sum_abs_vw'].mean(); print(f"  Avg Turnover (Long Only): EW={turnover_ew_long*100:.2f}%, VW={turnover_vw_long*100:.2f}%")
        long_monthly_agg = monthly_agg_data.get(9, pd.DataFrame())
        if not long_monthly_agg.empty:
            maxDD_ew_long=MDD(long_monthly_agg["excess_return_portfolio_ew"]); maxDD_vw_long=MDD(long_monthly_agg["excess_return_portfolio_vw"]); print(f"  Max Drawdown (Long Only): EW={abs(maxDD_ew_long)*100:.2f}%, VW={abs(maxDD_vw_long)*100:.2f}%")
            max_loss_ew_long=long_monthly_agg["excess_return_portfolio_ew"].min(); max_loss_vw_long=long_monthly_agg["excess_return_portfolio_vw"].min(); print(f"  Max 1M Loss (Long Only): EW={abs(max_loss_ew_long)*100:.2f}%, VW={abs(max_loss_vw_long)*100:.2f}%")
            if bench_data_loaded and 'bench_zeronet' in locals(): # Reuse benchmark data
                 bench_long=pd.merge(bench_zeronet[['MonthYear','monthly_return_bench','monthly_excess_return_bench']], long_monthly_agg, on='MonthYear',how='inner').dropna()
                 if not bench_long.empty:
                     long_res = results_summary_df.loc[9] if 9 in results_summary_df.index else pd.Series(dtype=float)
                     active_ret_ew_long=bench_long["excess_return_portfolio_ew"]-bench_long["monthly_excess_return_bench"]; active_ret_vw_long=bench_long["excess_return_portfolio_vw"]-bench_long["monthly_excess_return_bench"]
                     mean_active_ret_ew_long=active_ret_ew_long.mean(); mean_active_ret_vw_long=active_ret_vw_long.mean()
                     ir_ew_long=(mean_active_ret_ew_long/active_ret_ew_long.std())*np.sqrt(12) if active_ret_ew_long.std()>1e-9 else np.nan; ir_vw_long=(mean_active_ret_vw_long/active_ret_vw_long.std())*np.sqrt(12) if active_ret_vw_long.std()>1e-9 else np.nan
                     std_bench_raw_pct = bench_long["monthly_return_bench"].std()*100
                     index_perf = ["Mean Excess Return [%]",'Std RAW Return [%]', "Ann. Sharpe Ratio", "Max Drawdown [%]","Avg Monthly Turnover [%]", "Mean Active Return","Ann. Information Ratio"]
                     ew_long_chart=pd.DataFrame({benchmark_name:[mean_ret_bench_pct,std_bench_raw_pct,sr_bench,mdd_bench_pct,0,0,0],f'{model_name_label} Long':[long_res.get('ew_mean_ret',np.nan)*100,long_res.get('std_ew_ret',np.nan)*100,long_res.get('sharpe_ew',np.nan),abs(maxDD_ew_long)*100,turnover_ew_long*100,mean_active_ret_ew_long*100,ir_ew_long]},index=index_perf)
                     vw_long_chart=pd.DataFrame({benchmark_name:[mean_ret_bench_pct,std_bench_raw_pct,sr_bench,mdd_bench_pct,0,0,0],f'{model_name_label} Long':[long_res.get('vw_mean_ret',np.nan)*100,long_res.get('std_vw_ret',np.nan)*100,long_res.get('sharpe_vw',np.nan),abs(maxDD_vw_long)*100,turnover_vw_long*100,mean_active_ret_vw_long*100,ir_vw_long]},index=index_perf)
                     print(f"\n--- Risk-Adjusted Performance (Long Only vs {benchmark_name}, EW) ---"); print(ew_long_chart.round(3)); print(f"\n--- Risk-Adjusted Performance (Long Only vs {benchmark_name}, VW) ---"); print(vw_long_chart.round(3))
    else: print("  Ingen data for Long-Only portefølje.")

    print("\n--- Genererer kumulative avkastningsplott ---")
    plot_data = {}; # Prepare data for plots
    for j in [0, 9]: # Bottom and Top deciles
        if j in monthly_agg_data and not monthly_agg_data[j].empty:
            df_agg=monthly_agg_data[j].set_index('MonthYear').sort_index().shift(1).fillna(0);
            if not df_agg.empty: df_agg[f'cum_ret_ew_{j}']=(1+df_agg['excess_return_portfolio_ew']).cumprod()-1; df_agg[f'cum_ret_vw_{j}']=(1+df_agg['excess_return_portfolio_vw']).cumprod()-1; plot_data[j]=df_agg
    if bench_data_loaded and 'bench_zeronet' in locals(): bench_plot=bench_zeronet.set_index('MonthYear').sort_index().shift(1).fillna(0); bench_plot['cum_ret_bench']=(1+bench_plot['monthly_excess_return_bench']).cumprod()-1; plot_data['benchmark']=bench_plot
    # Plot EW
    fig_ew, ax_ew = plt.subplots(figsize=(15, 7)); legend_items=[]
    if 9 in plot_data: plot_data[9][f'cum_ret_ew_9'].plot(ax=ax_ew, label=f'{model_name_label} Long EW'); legend_items.append(f'{model_name_label} Long EW')
    if 0 in plot_data: plot_data[0][f'cum_ret_ew_0'].plot(ax=ax_ew, label=f'{model_name_label} Short EW'); legend_items.append(f'{model_name_label} Short EW')
    if 'benchmark' in plot_data: plot_data['benchmark']['cum_ret_bench'].plot(ax=ax_ew, label=benchmark_name, linestyle='--'); legend_items.append(benchmark_name)
    if legend_items: ax_ew.set_title(f'Kumulativ Excess Avkastning (EW & {benchmark_name})'); ax_ew.set_ylabel('Kumulativ Avkastning'); ax_ew.set_xlabel('Måned'); ax_ew.legend(legend_items); ax_ew.grid(True); fig_ew.tight_layout(); plt.show()
    else: plt.close(fig_ew); print("Ingen EW plot data.")
    # Plot VW
    fig_vw, ax_vw = plt.subplots(figsize=(15, 7)); legend_items=[]
    if 9 in plot_data: plot_data[9][f'cum_ret_vw_9'].plot(ax=ax_vw, label=f'{model_name_label} Long VW'); legend_items.append(f'{model_name_label} Long VW')
    if 0 in plot_data: plot_data[0][f'cum_ret_vw_0'].plot(ax=ax_vw, label=f'{model_name_label} Short VW'); legend_items.append(f'{model_name_label} Short VW')
    if 'benchmark' in plot_data: plot_data['benchmark']['cum_ret_bench'].plot(ax=ax_vw, label=benchmark_name, linestyle='--'); legend_items.append(benchmark_name)
    if legend_items: ax_vw.set_title(f'Kumulativ Excess Avkastning (VW & {benchmark_name})'); ax_vw.set_ylabel('Kumulativ Avkastning'); ax_vw.set_xlabel('Måned'); ax_vw.legend(legend_items); ax_vw.grid(True); fig_vw.tight_layout(); plt.show()
    else: plt.close(fig_vw); print("Ingen VW plot data.")

    print("--- Detaljert Porteføljeanalyse Fullført ---")
    return ew_table, vw_table, ew_chart, vw_chart, ew_long_chart, vw_long_chart # Return tables


# -----------------------------------------------------------------------------
# Step 6.7: Variable Importance Function (MODIFIED for GLM-H)
# -----------------------------------------------------------------------------
def calculate_variable_importance(df_standardized, features, target_col, train_idx, val_idx, param_grid=None):
    """ Calculates permutation importance for GLM-H using data from the provided indices. """
    print("\n--- Starter Variabel Viktighet Analyse (GLM-H, Siste Vindu) ---")
    model_name = "GLM-H" #<-- Changed Model Name
    if train_idx is None or val_idx is None or train_idx.empty or val_idx.empty: print("  FEIL: Ugyldige VI indekser."); return pd.DataFrame()
    print(f"  Bruker Train index ({len(train_idx)} obs), Val index ({len(val_idx)} obs)")
    X_train = df_standardized.loc[train_idx, features].values; y_train = df_standardized.loc[train_idx, target_col].values
    X_val = df_standardized.loc[val_idx, features].values; y_val = df_standardized.loc[val_idx, target_col].values
    X_train_val = np.vstack((X_train, X_val)); y_train_val = np.concatenate((y_train, y_val))
    if X_train_val.shape[0] < 2 or y_train_val.var() < 1e-15: print("  FEIL: VI data/varians."); return pd.DataFrame()
    ss_tot = np.sum((y_train_val - y_train_val.mean())**2)
    if ss_tot < 1e-15: print("  ADVARSEL: VI SS_tot er null."); return pd.DataFrame()

    # --- Tune and Train Base GLM-H Model ---
    print(f"  Tuner/trener base {model_name} modell for VI...")
    optim_param_vi = None; best_mse_vi = np.inf

    # *** Use the SAME param_grid definition as in run_glm_huber_on_window ***
    if param_grid is None: # Default GLM-H VI Grid
         param_grid = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0],
                       'epsilon': [1.1, 1.35, 1.5, 1.75, 2.0]}
                       # 'max_iter': [300]
    grid_vi = list(ParameterGrid(param_grid))

    for i, params in enumerate(grid_vi): # Tuning loop for VI base model
        try:
            # Instantiate HuberRegressor for tuning
            glm_val_vi = HuberRegressor(fit_intercept=True, **params, max_iter=300)
            glm_val_vi.fit(X_train, y_train)
            y_val_pred_vi = glm_val_vi.predict(X_val); current_mse_vi = mean_squared_error(y_val,y_val_pred_vi)
            if not np.isnan(current_mse_vi) and current_mse_vi < best_mse_vi: best_mse_vi=current_mse_vi; optim_param_vi=params
        except Exception: continue
    if optim_param_vi is None: print(f"  FEIL: VI tuning ({model_name}) feilet."); return pd.DataFrame()

    try: # Train base GLM-H model on Train+Val
        # Instantiate final base HuberRegressor
        base_model = HuberRegressor(fit_intercept=True, **optim_param_vi, max_iter=300)
        base_model.fit(X_train_val,y_train_val)
        base_preds = base_model.predict(X_train_val); ss_res_base = np.sum((y_train_val-base_preds)**2); base_r2 = 1-(ss_res_base/ss_tot)
        print(f"  Base R2 ({model_name}, siste Train+Val): {base_r2:.4f}")
    except Exception as e: print(f"  FEIL VI base modell ({model_name}): {e}"); return pd.DataFrame()

    # --- Permutation Importance (Zeroing out & Retraining) ---
    # This uses the retraining method for consistency with other models
    print("  Beregner viktighet (zeroing out & retraining)...")
    importance_results = {}
    for feature_idx, feature_name in enumerate(features):
        X_train_val_permuted = X_train_val.copy(); X_train_val_permuted[:, feature_idx] = 0
        try: # Retrain GLM-H with zeroed feature using same optimal params
            # Instantiate permuted HuberRegressor
            permuted_model = HuberRegressor(fit_intercept=True, **optim_param_vi, max_iter=300)
            permuted_model.fit(X_train_val_permuted, y_train_val)
            permuted_preds = permuted_model.predict(X_train_val_permuted)
            ss_res_permuted = np.sum((y_train_val - permuted_preds)**2)
            importance_results[feature_name] = 1 - (ss_res_permuted / ss_tot)
        except Exception: importance_results[feature_name] = base_r2 # Assign base R2 on error

    if not importance_results: print("  Ingen VI resultater."); return pd.DataFrame()
    imp_df = pd.DataFrame(importance_results.items(), columns=['Feature', 'R2_permuted'])
    imp_df['R2_reduction'] = (base_r2 - imp_df['R2_permuted']).clip(lower=0)
    total_reduction = imp_df['R2_reduction'].sum()
    imp_df['Importance'] = imp_df['R2_reduction']/total_reduction if total_reduction > 1e-9 else 0.0
    imp_df = imp_df.sort_values(by='Importance', ascending=False).reset_index(drop=True)
    print(f"\n--- Variabel Viktighet ({model_name}, Top 10) ---"); print(imp_df[['Feature', 'Importance']].head(10).round(4))
    # Plotting
    plt.figure(figsize=(12, 8)); top_n = 20
    plot_df = imp_df.head(top_n).sort_values(by='Importance', ascending=True)
    plt.barh(plot_df['Feature'], plot_df['Importance']); plt.xlabel("Relativ Viktighet (Permutation)"); plt.ylabel("Feature"); plt.title(f'{model_name} Variabel Viktighet (Top {top_n}) - Siste Vindu'); plt.tight_layout(); plt.show()
    return imp_df[['Feature', 'Importance']]

#-------------------------------------------------
# Step 6.8 Plot Time-Varying Complexity (Adapted for Alpha)
#-------------------------------------------------
def plot_time_varying_complexity(model_metrics, model_name='GLM-H'): #<-- Default Name Change
     """ Plots the optimal alpha over time """
     print(f"\n--- Plotter Tidsvarierende Modellkompleksitet (Optimal Alpha for {model_name}) ---") #<-- Title Change
     complexity_param = 'optim_alpha' # Parameter to plot
     if model_name in model_metrics and complexity_param in model_metrics[model_name]:
         alphas = model_metrics[model_name][complexity_param]
         if alphas and not all(np.isnan(d) for d in alphas):
             valid_alphas = [(i+1, a) for i, a in enumerate(alphas) if pd.notna(a)]
             if valid_alphas:
                 windows, plot_alphas = zip(*valid_alphas)
                 alpha_data = pd.DataFrame({complexity_param: plot_alphas}, index=pd.Index(windows, name='Window'))
                 print(f"\n--- Optimal {complexity_param} per Vindu Tabell ({model_name}) ---"); print(alpha_data) #<-- Title Change
                 # Plotting on log scale can be useful for alpha
                 plt.figure(figsize=(10, 5)); plt.plot(windows, plot_alphas, marker='o', linestyle='-')
                 plt.yscale('log') # Use log scale for alpha
                 plt.xlabel("Rullerende Vindu Nummer"); plt.ylabel(f"Optimal {complexity_param} funnet (Log Scale)") #<-- Label Change
                 plt.title(f"{model_name} Optimal {complexity_param} per Vindu"); plt.grid(True); plt.tight_layout(); plt.show() #<-- Title Change
             else: print(f"  Ingen gyldige {complexity_param} verdier.")
         else: print(f"  Ingen {complexity_param} data.")
     else: print(f"  Ingen '{complexity_param}' data for {model_name}.")


# ------------------------------------------------
# Step 8: Main Orchestration Function (Uses GLM-H now)
# ------------------------------------------------
def run_analysis_for_subset(file_path, data_subset='all', benchmark_file=None, filter_portfolio_construction=False):
    """ Runs the full pipeline for one data subset using HuberRegressor (GLM-H). """
    run_label = data_subset.capitalize(); model_name = 'GLM-H' #<-- Set Model Name
    start_time = datetime.datetime.now()
    print(f"\n{'='*20} Starter Kjøring: {model_name} {run_label} Firms {'='*20}")
    print(f"Starttid: {start_time.strftime('%Y-%m-%d %H:%M:%S')}")

    # --- Load & Subset ---
    print("Laster rådata...")
    df_raw = load_prepare_data(file_path)
    if df_raw is None: return np.nan, None, None, (None,)*6
    if 'MarketCap_orig' not in df_raw.columns: print("FEIL: MarketCap_orig mangler."); return np.nan, None, None, (None,)*6
    print(f"Lager subset: {run_label}...")
    df = pd.DataFrame(); df_raw_mc = df_raw.dropna(subset=['MarketCap_orig', 'Date'])
    if data_subset == 'all': df = df_raw.copy()
    elif data_subset == 'big': p=0.3; df = df_raw_mc.groupby('Date', group_keys=False).apply(lambda x: x.nlargest(int(len(x)*p),"MarketCap_orig"))
    elif data_subset == 'small': p=0.3; df = df_raw_mc.groupby('Date', group_keys=False).apply(lambda x: x.nsmallest(max(1,int(len(x)*p)),"MarketCap_orig"))
    else: print(f"FEIL: Ukjent subset '{data_subset}'."); return np.nan, None, None, (None,)*6
    if df.empty: print(f"FEIL: Tomt subset '{run_label}'."); return np.nan, None, None, (None,)*6
    print(f"Subset '{run_label}' form: {df.shape}")

    # --- Standardize & Clean ---
    glm_features = define_features(df) # Use the auto-detected features
    if not glm_features: print("FEIL: Ingen features i subset."); return np.nan, None, None, (None,)*6
    df = rank_standardize_features(df, glm_features)
    essential_cols = ['TargetReturn_t', 'NextMonthlyReturn_t+1', 'MarketCap_orig', 'Date', 'Instrument']
    df = clean_data(df, glm_features, essential_cols, target="TargetReturn_t")
    if df.empty: print("FEIL: Tomt etter rensing."); return np.nan, None, None, (None,)*6
    glm_features = [f for f in glm_features if f in df.columns] # Refresh
    if not glm_features: print("FEIL: Ingen features igjen."); return np.nan, None, None, (None,)*6
    df = df.sort_values("Date").reset_index(drop=True)

    # --- Rolling Window ---
    train_period = 108; val_period = 72; test_period = 12 # Your splits
    print(f"\n--- Rullerende Vindu ({run_label}, {model_name}, Train={train_period}, Val={val_period}, Test={test_period}) ---")
    results_list = []; model_metrics = defaultdict(lambda: defaultdict(list))
    yhat_col_name = f'yhat_{model_name.lower().replace("-", "_")}' # e.g., yhat_glm_h
    last_train_idx, last_val_idx = None, None

    try: # Add/remove temporary MonthYear column
        needs_monthyear = 'MonthYear' not in df.columns
        if needs_monthyear: df['MonthYear'] = df['Date'].dt.to_period('M')
        splits_generator = get_rolling_splits(df, train_period, val_period, test_period)
        splits = list(splits_generator)
        if needs_monthyear: df.drop(columns=['MonthYear'], inplace=True)
        num_windows = len(splits)
    except ValueError as e: print(f"FEIL ved splits: {e}\nAvslutter {run_label}."); return np.nan, None, None, (None,)*6
    print(f"Antall vinduer: {num_windows}\n");
    if num_windows == 0: print("Ingen vinduer. Avslutter."); return np.nan, None, None, (None,)*6

    # --- Rolling Window Loop ---
    for window, (train_idx, val_idx, test_idx, train_dates, val_dates, test_dates) in enumerate(splits):
        window_start_time = datetime.datetime.now()
        test_start_str = test_dates['min'].strftime('%Y-%m-%d') if test_dates is not None else "N/A"
        if test_idx.empty or val_idx.empty or train_idx.empty: print(f"Vindu {window+1}: Tomt train/val/test. Hopper over."); continue
        print(f"Vindu {window + 1}/{num_windows}: Train {train_dates['min'].strftime('%Y-%m-%d')} - {train_dates['max'].strftime('%Y-%m-%d')} ({len(train_idx)} obs), "
              f"Val {val_dates['min'].strftime('%Y-%m-%d')} - {val_dates['max'].strftime('%Y-%m-%d')} ({len(val_idx)} obs), "
              f"Test {test_dates['min'].strftime('%Y-%m-%d')} - {test_dates['max'].strftime('%Y-%m-%d')} ({len(test_idx)} obs)")

        X_train = df.loc[train_idx, glm_features].values; y_train = df.loc[train_idx, "TargetReturn_t"].values
        X_val = df.loc[val_idx, glm_features].values; y_val = df.loc[val_idx, "TargetReturn_t"].values
        X_test = df.loc[test_idx, glm_features].values; y_test = df.loc[test_idx, "TargetReturn_t"].values
        if np.isnan(y_train).all() or np.isnan(y_val).all() or X_train.shape[0]<2 or X_val.shape[0]<2: print("  Data mangler/NaNs. Hopper."); continue

        # *** Call the run_glm_huber_on_window function ***
        preds_oos, r2_oos, _, _, _, _, _, optim_alpha = run_glm_huber_on_window(
            X_train, y_train, X_val, y_val, X_test, y_test
            # Pass param_grid=... here if you want to override the default
        )

        window_predictions = {'Date': df.loc[test_idx, 'Date'].values, 'Instrument': df.loc[test_idx, 'Instrument'].values, 'TargetReturn_t': y_test, yhat_col_name: preds_oos}
        results_list.append(pd.DataFrame(window_predictions))
        model_metrics[model_name]['oos_r2'].append(r2_oos) # Store window R2
        model_metrics[model_name]['optim_alpha'].append(optim_alpha) # Store optimal alpha
        last_train_idx, last_val_idx = train_idx.copy(), val_idx.copy() # Store indices for VI
        print(f"  Vindu {window + 1} fullført på {(datetime.datetime.now() - window_start_time).total_seconds():.1f}s. Optimal Alpha: {optim_alpha:.4f}, Window OOS R2: {r2_oos:.4f}")

    # --- Aggregate & Analyze ---
    if not results_list: print(f"\nIngen resultater for {run_label}."); return np.nan, None, model_metrics, (None,)*6
    results_df = pd.concat(results_list).reset_index(drop=True)
    print(f"\n--- Samlet Resultatanalyse ({run_label}, {model_name}) ---")
    y_true=results_df['TargetReturn_t']; y_pred=results_df[yhat_col_name]; valid_idx=y_true.notna() & y_pred.notna()
    y_t_val=y_true[valid_idx]; y_p_val=y_pred[valid_idx]; R2OOS_overall=np.nan
    if len(y_t_val)>1 and y_t_val.var()>1e-15:
        ss_res=np.sum((y_t_val-y_p_val)**2); ss_true_sq=np.sum(y_t_val**2)
        if ss_true_sq>1e-15: R2OOS_overall=1-(ss_res/ss_true_sq)
    avg_window_r2 = np.nanmean(model_metrics[model_name]['oos_r2'])
    print(f"Overall OOS R² ({run_label}, Gu et al.): {R2OOS_overall:.6f}")
    print(f"Average Window OOS R² ({run_label}):    {avg_window_r2:.6f}")
    model_metrics[model_name]['oos_r2_overall_gu'] = R2OOS_overall

    # --- Detailed Portfolio Analysis ---
    portfolio_tables = perform_detailed_portfolio_analysis(
        results_df, df, benchmark_file=benchmark_file,
        filter_small_caps=filter_portfolio_construction,
        model_name_label=model_name # Pass model name label
    )

    # --- VI and Complexity Plots (Only for 'all' subset run) ---
    vi_df = None
    if data_subset == 'all':
        if last_train_idx is not None and last_val_idx is not None:
             # Use calculate_variable_importance (which now uses GLM-H)
             # Define the parameter grid for VI (should match the one used in rolling window)
             glm_param_grid = {'alpha': [0.0001, 0.001, 0.01, 0.1, 1.0, 10.0],
                               'epsilon': [1.1, 1.35, 1.5, 1.75, 2.0]}
             vi_df = calculate_variable_importance(df, glm_features, 'TargetReturn_t', last_train_idx, last_val_idx, param_grid=glm_param_grid)
        else: print("Kan ikke kjøre VI (mangler siste vindu indekser).")
        plot_time_varying_complexity(model_metrics, model_name) # Pass model name

    end_time = datetime.datetime.now()
    print(f"\n--- Kjøring ({run_label}, {model_name}) fullført: {end_time.strftime('%Y-%m-%d %H:%M:%S')} (Total tid: {(end_time - start_time)}) ---")
    print(f"{'='*70}")
    return R2OOS_overall, vi_df, model_metrics, portfolio_tables

# --------------------------------------------------------------------------
# Main Execution Block
# --------------------------------------------------------------------------
if __name__ == "__main__":
    # --- CONFIGURATION ---
    # *** SET YOUR FILE PATHS HERE ***
    data_file = "Cleaned_OSEFX_Market_Macro_Data.csv" # Or your specific file name
    # *** SET YOUR OSEBX BENCHMARK FILE PATH ***
    benchmark_csv_file = r"path/to/your/OSEBX_data.csv" # <--- UPDATE THIS LINE

    if not os.path.exists(data_file):
        print(f"FEIL: Input datafil ikke funnet på '{data_file}'. Avslutter.")
        exit()
    if benchmark_csv_file and not os.path.exists(benchmark_csv_file):
        print(f"ADVARSEL: Benchmark fil ikke funnet på '{benchmark_csv_file}'. Fortsetter uten.")
        benchmark_csv_file = None

    # Filter small caps during portfolio construction?
    filter_small_caps_portfolio = False

    # --- Optional: Define directory for saving results ---
    output_dir = "GLM_H_Results" #<-- Changed Directory Name
    if not os.path.exists(output_dir):
        os.makedirs(output_dir); print(f"Opprettet mappe: {output_dir}")

    # --- RUN ANALYSIS FOR EACH SUBSET ---
    results_r2_summary = {}
    variable_importance_df = None
    all_metrics = {}
    portfolio_results_tables = {}

    for subset in ['all', 'big', 'small']:
        r2, vi_df_run, metrics_run, port_tables_run = run_analysis_for_subset(
            file_path=data_file,
            data_subset=subset,
            benchmark_file=benchmark_csv_file,
            filter_portfolio_construction=filter_small_caps_portfolio
        )
        results_r2_summary[subset] = r2
        all_metrics[subset] = metrics_run
        portfolio_results_tables[subset] = port_tables_run
        if subset == 'all' and vi_df_run is not None:
             variable_importance_df = vi_df_run

    # --- Print Final R2 Summary Table ---
    print("\n\n" + "="*30 + " Final OOS R2 Summary (GLM-H) " + "="*30) #<-- Title Change
    summary_r2_data = {
        "Full sample": [results_r2_summary.get('all', np.nan)],
        "Large firms": [results_r2_summary.get('big', np.nan)],
        "Small firms": [results_r2_summary.get('small', np.nan)]
    }
    r2_summary_df = pd.DataFrame.from_dict(summary_r2_data, orient='index', columns=['GLM-H R2OOS (Gu et al.)']) #<-- Column Name Change
    print(r2_summary_df.round(6))
    try: # Save summary table
        r2_summary_df.to_csv(os.path.join(output_dir, "glm_h_r2_summary_subsets.csv")) #<-- Filename Change
        print(f" -> R2 Sammendrag lagret til {output_dir}")
    except Exception as e: print(f"  FEIL ved lagring av R2 Sammendrag: {e}")
    print("="*78)

    # Save VI results if calculated
    if variable_importance_df is not None:
         print("\nLagrer Variabel Viktighet resultater...")
         try:
              variable_importance_df.to_csv(os.path.join(output_dir, "glm_h_variable_importance_last_window.csv"), index=False) #<-- Filename Change
              print(f" -> Variabel Viktighet lagret til {output_dir}")
         except Exception as e: print(f"  FEIL ved lagring av VI: {e}")

    # Save portfolio tables if generated
    for subset, tables in portfolio_results_tables.items():
         if tables and len(tables) == 6:
              names = ['decile_ew', 'decile_vw', 'hl_vs_bench_ew', 'hl_vs_bench_vw', 'long_vs_bench_ew', 'long_vs_bench_vw']
              for i, table_df in enumerate(tables):
                   if table_df is not None and not table_df.empty:
                        filename = os.path.join(output_dir, f"glm_h_portfolio_{subset}_{names[i]}.csv") #<-- Prefix Change
                        try: table_df.to_csv(filename); print(f" -> Porteføljetabell lagret til {filename}")
                        except Exception as e: print(f"  FEIL ved lagring av porteføljetabell {filename}: {e}")

    print("\nFull GLM-H analyse fullført.")
```
